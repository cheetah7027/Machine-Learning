```python
#fatorial Function
def fatorial(n):
    fact=1
    fact=fact*n
    for i in range(1,n):
        fact=fact*(n-i)
    return fact
print(fatorial(5))
```

```python
def factorial_num(n):
  fact=1
  for i in range(1,n+1):
    fact=fact*n;
    return fact
n=int(input('enter the number'))
print(factorial_num(n))
```

Stemming is the process of reducing a word to its keyword

```python
port_stem=PorterStemmer()
```

```python
def stemming(content):
  stemmed_content=re.sub('[^a-zA-Z]','',content)
  stemmed_content=stemmed_content.lower()
  stemmed_content=stemmed_content.split()
  stemmed_content=[port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content=' '.join(stemmed_content)
  return stemmed_content
```

```python
news_data=news_data['content'].apply(stemming)
```

Vectos **Operation**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Dot Porduct of two vectors


```python
a=np.array([2,3])
b=np.array([-7,1])
a_dot_b=np.dot(a,b)
print(a_dot_b)
```

Cross Porduct of two vectors

```python
a=np.array([1,2,3])
b=np.array([4,-7,1])
a_cross_b=np.cross(a,b)
print(a_cross_b)
```

Projection of the one vector on another vector

```python
a=np.array([2,3])
b=np.array([-7,1])
#magnitude of b vector
mag_b=np.sqrt(sum(b**2))

projection=(np.dot(a,b)/mag_b**2)*b
print(projection)
```

```python
a=np.array([4,2,3])
b=np.array([-9,-7,1])
#magnitude of b vector
mag_b=np.sqrt(sum(b**2))

projection=(np.dot(a,b)/mag_b**2)*b
print(projection)
```
